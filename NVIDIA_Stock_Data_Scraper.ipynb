```python
#import libraries
from alpha_vantage.timeseries import TimeSeries

import requests
from bs4 import BeautifulSoup

import pandas as pd
import io
```

```python
#build TimeSeries variable from Alpha Vantage API
ts1 = TimeSeries(key = API_key)
```

### Challenge: Analyze NVDA Stock

```python
#get monthly data of NVDA using the alpha_vantage package
ts1.get_monthly('NVDA')
```

```python
#get weekly data of NVDA using the alpha_vantage package
ts1.get_weekly('NVDA')
```

```python
#pull income statement for NVDA with requests package
url = 'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol=NVDA&apikey=' + str(API_key)
r = requests.get(url)
fd = BeautifulSoup(r.content)

print(fd)
```

```python
#get Bollinger bands values for NVIDIA with requests package, daily interval, 10 time periods, open stock value
url = 'https://www.alphavantage.co/query?function=BBANDS&symbol=NVDA&interval=daily&time_period=10&series_type=open&nbdevup=3&nbdevdn=3&apikey=' + str(API_key)
r = requests.get(url)
ti = BeautifulSoup(r.content)

print(ti)
```
